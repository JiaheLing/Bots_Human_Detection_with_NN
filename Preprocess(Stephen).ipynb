```python
import numpy as np
import pandas as pd
# nltk.download()
```

```python
data = pd.read_csv("./Bots_Human_Detection_with_NN/MarchtoMay_TweetsandNews_fortopicmodeling.csv", dtype=str)
```

```python
df = data[['Handles', 'Content', 'SourceatBot7']]
```

### Tokenization

```python
from nltk.tokenize import sent_tokenize, word_tokenize
# N = 100
# sample = df.head(N)
# sentences = sample[["Content"]].head(N)
N = len(df)
sample = df
sentences = sample[["Content"]]
# https://stackoverflow.com/questions/70410115/how-to-access-items-from-ndarray-with-dtype-object
tw = np.empty(N, dtype=object)
ts = np.empty(N, dtype=object)
for i in range(len(sentences)):
    sen = sentences.iloc[i].Content
    word_tk = word_tokenize(sen)
    tw[i] = word_tk
    sen_tk = sent_tokenize(sen)
    ts[i] = sen_tk

sample.insert(sample.shape[1], 'Tokenize(W)', tw)
#sample.insert(sample.shape[1], 'Tokenize(S)', ts)
```

```python
sample
```

### Stopwords & Punctuation Removal

```python
import nltk
import string
from nltk.corpus import stopwords
nltk.download('stopwords', quiet=True)

def remove_stopwords(input_text):
    return [token for token in input_text if token.lower() not in stopwords.words('english')]

def remove_punctuation(input_text):
    return [token for token in input_text if token not in set(string.punctuation)]
```

```python
t_sw = np.empty(N, dtype=object)
t_pc = np.empty(N, dtype=object)
# stopwords
for i in range(len(sentences)):
    # stopwords
    sw_remove = remove_stopwords(sample['Tokenize(W)'][i])
    t_sw[i] = sw_remove
sample.insert(sample.shape[1], 'Tokenize_Stopword', t_sw)
# punctuation
for i in range(len(sentences)):
    # stopwords
    pc_remove = remove_punctuation(sample['Tokenize_Stopword'][i])
    t_pc[i] = pc_remove
sample.insert(sample.shape[1], 'Token_Stop_Punc', t_pc)
```

```python
sample
```

### Stemming

Some people prefer lemmatization since it is less aggressive and the words still are valid. BUt stemming is used here.

```python
from nltk.stem import SnowballStemmer
st = np.empty(N, dtype=object)

def stem(input_text):
    stemmer = SnowballStemmer('english')
    return [stemmer.stem(token) for token in input_text]

for i in range(len(sentences)):
    steming = stem(sample['Token_Stop_Punc'][i])
    st[i] = steming
sample.insert(sample.shape[1], 'Stemmed', st)

# Apply lemmatize function
print(stem(sample['Token_Stop_Punc'][1]))
print(sample['Token_Stop_Punc'][1])
```

```python
sen_final = np.empty(N, dtype=object)
```

```python
for i in range(len(sen_final)):
    sen_final[i] = ' '.join(sample["Stemmed"][i])
sample.insert(sample.shape[1], 'Processed', sen_final)
```

```python
df = sample[['Handles', 'Stemmed', 'SourceatBot7', 'Processed']]
df.rename(columns = {'Processed':'Content'}, inplace=True)
df
```
